## 1. Two Sum https://leetcode.com/problems/two-sum/description/

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]

Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]


```python

def find_two_sum(nums: List[int], target: int):
    # constuct a value index map {2: 1, 8: 2, 6: 3, 5: 4} value and its Index
    index_val_map = {}
    count = 1
    for n in nums:
        # diff target - current iteration
        diff = target - n
        print('<<< diff>>>', diff)
        if diff in index_val_map:
            # here if diffrence one found just return the list
            return [index_val_map[diff], count]
        
        index_val_map[n] = count
        print(">>>>>> index_val_map", index_val_map)
        count += 1
    
```

```python
find_two_sum([2,8,6,5,7,11,15], 9)
```

## 2. Duplicate find
https://leetcode.com/problems/contains-duplicate/

Input: nums = [1, 2, 3, 3] if it has dublicate return True else False

```python
def has_duplicate(nums: List[int]) -> bool:
    value_count = {}
    for n in nums:
        if n in value_count:
            return True
        else:
            value_count[n] = 1
    return False
```

```python
has_duplicate( [1, 2, 3, 3])
```

```python
has_duplicate( [1, 2, 3, 4])
```

##  3. isAnagram
https://leetcode.com/problems/valid-anagram/

Input: s = "racecar", t = "carrace"

Output: true

if same char can make another string

```python
def is_anagram(string_one: str, string_two:str) -> bool:
    if len(string_one) != len(string_two):
        return False

    count_string_one = {}
    count_string_two = {}
    for i in range(len(string_one)):
        count_string_one[string_one[i]] = 1 + count_string_one.get(string_one[i], 0)
        count_string_two[string_two[i]] = 1 + count_string_two.get(string_two[i], 0)

    return count_string_one == count_string_two

        
```

```python
is_anagram("racecar", "carrace")
```

```python
is_anagram("jam", "jar")
```

## 4. Valid Palindrome 
https://leetcode.com/problems/valid-palindrome/description/

Input: s = "Was it a car or a cat I saw?" -> True

Input: s = "tab a cat"  -> False



```python
def is_palindrome(s):
    clean_string = ""
    for c in s:
        if c.isalnum():
            clean_string += c.lower()
    return clean_string == clean_string[::-1]
```

```python
is_palindrome("tab a cat")
```

```python
is_palindrome("Was it a car or a cat I saw?")
```

## 5. Valid Parentheses

https://leetcode.com/problems/valid-parentheses/description/

Example 1:

Input: s = "()"

Output: true

Example 2:

Input: s = "()[]{}"

Output: true


```python
def is_valid_parentheses(s: str) -> bool:
    stack = []
    close_to_open_map = {
        ")":"(",
        "]":"[",
        "}":"{"
    }
    for c in s:
        if c in close_to_open_map:
            if stack and stack[-1] == close_to_open_map[c]:
                stack.pop()
            else:
                return False
        else:
            stack.append(c)
    if stack:
        return False

    return True
```

```python
is_valid_parentheses("()")
```

```python
is_valid_parentheses("()[]{}")
```

```python
is_valid_parentheses("([)]")
```
